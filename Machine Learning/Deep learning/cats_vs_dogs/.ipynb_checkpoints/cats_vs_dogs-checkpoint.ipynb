## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from imports_for_ML import load_images

from sklearn.model_selection import train_test_split

from tensorflow.keras.preprocessing.image import load_img, img_to_array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Conv2D, MaxPooling2D, Flatten, BatchNormalization
from keras.optimizers import Adamax
import keras
from tensorflow.keras.models import save_model
```

## Loading dataset

```python
img_dir = '/Users/konrad/Desktop/Python_projects/cats_vs_dogs/train'
imgs = os.listdir(img_dir)

imgs_path = [os.path.join(img_dir, img) for img in imgs]
# Loading pathnames of cats and dogs
```

## Building CNN

```python
X, y = load_images(img_dir, n_images=25000, resize=(50, 50))
# Loading 10000 photos from train folder in 50x50p size

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
# Creating training and testing variables
```

```python
cat_vs_dog = Sequential([
    
    
    Conv2D(32, (3,3), activation = 'relu' ),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    # 1st layer
    
    Conv2D(64,(3,3), activation = 'relu' ),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    # 2nd layer
    
    Conv2D(128,(3,3), activation = 'relu' ),
    BatchNormalization(),
    MaxPooling2D((2,2)),
    # 3rd layer

    
    Flatten(input_shape = (5, 5,32)),
    Dense(128, activation = 'relu'),
    BatchNormalization(),
    Dense(1, activation='sigmoid')
])
cat_vs_dog.compile(loss = 'binary_crossentropy', optimizer = Adamax(lr=0.01), metrics = ['accuracy'])
# Compiling the model
```

```python
history = cat_vs_dog.fit(X_train, y_train, epochs = 10, validation_data = (X_test, y_test))
```

```python
print('Accuracy on the training data:')
print(cat_vs_dog.evaluate(X_train, y_train)[1])

print('\nAccuracy on the test data:')
print(cat_vs_dog.evaluate(X_test, y_test)[1])
```

## Visualization

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
# Graph for model accuracy
```

```python
plt.plot(history.history['loss'], marker = 'o')
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
# Graph for model loss
```
