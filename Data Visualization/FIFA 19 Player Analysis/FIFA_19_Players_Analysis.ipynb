```python
feat = ['Name',
       'Age',
       'Overall',
       'Nationality',
       'Club', 'Value',
       'Crossing', 'Finishing',
       'HeadingAccuracy', 'ShortPassing',
       'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing',
       'BallControl', 'Acceleration', 'SprintSpeed',
       'Agility', 'Reactions', 'Balance',
       'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']
# Choosing some of the most important features for visualization
```

## Analyzing age
I will now check how age correlates with different features from this dataframe

```python
# Distribution of age among the players in dataset
sns.set(font_scale = 0.75)
sns.countplot(data=updated_players, x = 'Age')
plt.title('Distribution of age among players')
plt.ylabel('Number of players')
```

```python
age_feat = ['Stamina', 'Strength', 'Jumping', 'Agility', 'Acceleration', 'SprintSpeed', 'Reactions']

for feature in age_feat:
    plt.figure(figsize=(5, 5))
    sns.regplot(x = 'Age', y = feature, data = updated_players)
```

The most interesting plots are age-stamina where stamina increases as players get older, and jumping-age where players jump higher as they get older

### Overall rating vs Age
I am going to check how age affects the overall ratings of players

```python
plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
sns.regplot(data = updated_players, x = 'Age', y = 'Overall')
plt.title('Age vs Overall rating of football players')
plt.show()
```

### Age distribution in top teams

```python
teams = ['Real Madrid', 'FC Barcelona',
         'Paris Saint-Germain', 'Juventus',
         'AC Milan', 'Manchester United',
         'Manchester City', 'Liverpool',
         'FC Bayern München']
team_players = updated_players.loc[updated_players['Club'].isin(teams) & updated_players['Age']]

plt.figure(figsize=(20, 10))
sns.set(font_scale=1.5)
sns.violinplot(data= team_players, y = 'Age', x = 'Club', rotation=90)
plt.xticks(rotation=60)
plt.title('Distribution of team age among top teams')
plt.show()
```

### Analysing best clubs
In this part of the project I will try to find the best players and teams that have players with highest overall rating based on the data

```python
best_teams = {}

for team in updated_players['Club'].unique():
    rating = updated_players['Overall'][updated_players['Club'] == team].sum()
    best_teams[team] = rating
```

```python
print(type(best_teams))
# best_teams variable is a dict so I am going to convert it to DataFrame in order to create visualizations
best_teams = pd.DataFrame.from_dict(best_teams, orient= 'index', columns= ['Rating'])
best_teams['Club'] = best_teams.index

print(type(best_teams))
# I successfully changed the type of best_teams from dict to DataFrame
```

```python
best_teams = best_teams.sort_values(by='Rating', ascending=False).head(10)
# Leaving only top 10 teams to compare them
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(data = best_teams, x= 'Club', y='Rating')
plt.xticks(rotation=60)
plt.ylim(1500, 2600)
plt.show()
```

Based on the graph it can be concluded that Real Madrid and FC Barcelona had the best players overall in FIFA19, on the other side of the graph, the worst overall rating among top 10 clubs had Arsenal and Atletico Madrid.

Now I am going to find countries with the highest overall

### Analysing best countries
I will now do the same steps as in the best teams analysis but with 'Nationality' instead of 'Club' column

```python
best_countries = {}

for country in updated_players['Nationality'].unique():
    rating = updated_players['Overall'][updated_players['Nationality'] == country].sum()
    best_countries[country] = rating


best_countries = pd.DataFrame.from_dict(best_countries, orient= 'index', columns= ['Rating'])
best_countries['Nationality'] = best_countries.index

best_countries = best_countries.sort_values(by='Rating', ascending=False).head(10)
```

### Building the best team
I will now find what statistics are the most important for each position

```python
attributes = ['Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']
# Listing all the attributes players can have

players['Position'].unique()
# Checking what positions are given in the dataset
```

```python
best_position = players.groupby('Position')[attributes].mean()
# Creating a dataframe with mean values of attributes for each position
best_position
```

I am going to pick the 5 highest attributes for each position

```python
best_stats = pd.DataFrame()
for position, attribute in best_position.iterrows():
    top = dict(attribute.nlargest(n=5))
    best_stats[position] = tuple(top)
# Making a dataframe of the most important attributes for particular positions
best_stats
```

### Conclusion
The dataframe above represents the most important attributes for each position. Presumably, player whose sum of these attributes is the highest will be the best peroforming player on his position in FIFA19.
